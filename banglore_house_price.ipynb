```python
# path= path is fethched from documentation
path = r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB" 
banglore_house = pd.read_csv(path)

```

split data

```python
x=banglore_house[["bath","balcony","total_sqft_int","bhk","price_per_sqft"]]
y=banglore_house["price"]
```

```python
print("shape of x:",x.shape)
print("shape of y:",y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=51)
```

```python
print("shape of X_train:",x_train.shape)
print("shape of y_train:",y_train.shape)
print("shape of x_test:",x_test.shape)
print("shape of y_test:",y_test.shape)
```

FEATURE SCALLING

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
sc.fit(x_train) # fit x_train i.e independent training data
```

```python
x_train=sc.transform(x_train)
x_test=sc.transform(x_test) 
#transfrom is used to turn skew distribution into normal distribution
```

MODEL TRAINING := LINEAR REGRESSION


```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
#for building model fit x and y train data
```

```python
lr.coef_ # what coeefficient value our model had learnt i.e y=mx+c  coefficient = m here our euation will b i.e y=C+M1X1+M2X2+.......M107X107
```

```python
lr.intercept_ # gives the value of C
```

```python
predicted=pd.DataFrame(lr.predict(x_test))
predicted.head()
# it will give result on basis odf x_train.. manuallly it is not posible to five features and predict for million of data
```

PREDICT THE PRICE  OF HOUSE AND TEST

```python
lr.score(x_test,y_test)
#to chek the accuracy of model
```

split dataset

```python
x=banglore_house.drop("price",axis=1)
y=banglore_house["price"]
print("shape of x:",x.shape)
print("shape of y:",y.shape)
```

train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=51,test_size=0.2)
```

feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

svm model training

```python
from sklearn.svm import SVR
```

```python
svr_rbf=SVR(kernel="rbf")
svr_rbf.fit(x_train,y_train)
svr_rbf.predict(x_test)
svr_rbf.score(x_test,y_test)

#this kernel is not givin proper accuracy so we will go for another kernel
```

```python
svr_rbf=SVR(kernel="poly")
svr_rbf.fit(x_train,y_train)
svr_rbf.predict(x_test)
svr_rbf.score(x_test,y_test)
```

```python
svr_rbf=SVR(kernel="linear",degree=6)
svr_rbf.fit(x_train,y_train)

svr_rbf.score(x_test,y_test)
print(svr_rbf.predict(x_test))
```

```python
svr_rbf=SVR(kernel="sigmoid")
svr_rbf.fit(x_train,y_train)
svr_rbf.predict(x_test)
svr_rbf.score(x_test,y_test)
```
